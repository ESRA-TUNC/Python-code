```python
import matplotlib.pyplot as plt

kategoriler=['A','B','C','D','E']
değerler =[23,45,54,12,60]

plt.figure(figsize=(8,6))
plt.bar(kategoriler,değerler,color='skyblue')


plt.title('Kategori Değerlerini Karşılaştırma',fontsize=18)
plt.xlabel('Kategoriler',fontsize=14)
plt.ylabel('Değerler',fontsize=14)

plt.ylim(0,60)
plt.show()
```
