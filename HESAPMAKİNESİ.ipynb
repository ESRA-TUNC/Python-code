```python
#HESAP MAKİNESİ UYGULAMASI
import datetime
import pytz

#fonksiyon tanımlamaları
def toplama(x,y):
      return x+y

def çıkartma(x,y):
      return x-y
def çarpma (x,y):
      return x*y
def bolme (x,y):
   if y == 0:
      return "hata: sıfıra bölünemez..."
   return x/y

#dosya oluşturur ve işlemleri ekler.
def dosyaya_yaz(sayi1,sayi2,operator,sonuc):
    with open("hesap geçmişi.txt","a",encoding="utf-8") as dosya:
      local_tz=pytz.timezone('europe/istanbul')      # zaman dilimini okur
      zaman=datetime.datetime.now(local_tz).strftime("%Y-%d-%m  %H:%M:%S")
      dosya.write(f"{zaman} ---> {sayi1} {operator} {sayi2}={sonuc}\n")

#oluşturduğumuz dosyayı olır.
def dosyadan_oku():
    with open("hesap geçmişi.txt","r",encoding="utf-8") as dosya:
      geçmis=dosya.read()
      print("-----------HESAP GEÇMİŞİ--------")
      print(geçmis)

#  ANA AKIŞ FONKSİYONU
def hesap_makinesi():
    while True:
      print("\nhesap makinesi uygulamasına hoşgeldiniz...")
      print("istediğiniz uygulamayı seçiniz..")
      print("1.toplama işlemi")
      print("2.çıkartma işlemi")
      print("3.çarpma işlemi")
      print("4.bölme işlemi")
      print("5.hesap geçmişini görüntüle")
      print("6.'q' çıkış...")

      secim = input("seçiminiz(1/2/3/4/5/q): ")

      if secim =="q":
            print("çıkış yapılıyor...")
            break    #  while döngüsü ile program sonlanır.
      elif  secim == "5":
            dosyadan_oku()

      elif secim in ['1','2','3','4']:
            try:                         #try ile hata oluşacak kısım
              sayi1=float(input("birinci sayıyı giriniz: "))
              sayi2=float(input("ikinci sayıyı giriniz: "))
            except:                      #hata olur ise ne yapılacağı
              print("hatalı giriş yaptınız. ")
              continue                   #hatalı giriş oldu başa dön demektir.
            if secim == '1':
               sonuc=toplama(sayi1,sayi2)
               print(f"{sayi1}+{sayi2}={sonuc}")
               operator="+"              # dosyaya işlemleri ekler. bu 2 satır.
               dosyaya_yaz(sayi1,sayi2,operator,sonuc)

            elif secim == '2':
               sonuc=çıkartma(sayi1,sayi2)
               print(f"{sayi1}-{sayi2}={sonuc}")
               operator="-"
               dosyaya_yaz(sayi1,sayi2,operator,sonuc)

            elif secim == '3':
               sonuc=çarpma(sayi1,sayi2)
               print(f"{sayi1}x{sayi2}={sonuc}")
               operator="*"
               dosyaya_yaz(sayi1,sayi2,operator,sonuc)

            if secim == '4':
               sonuc=bolme(sayi1,sayi2)
               print(f"{sayi1}/{sayi2}={sonuc}")
               operator="/"
               dosyaya_yaz(sayi1,sayi2,operator,sonuc)

      else:
        print("geçersiz giriş 1,2,3,4,5 veya q giriniz...")

#tanımladığımız fonksiyonu çağırdık.
hesap_makinesi()

```
